# Alzheimer Dataset Exploratory Analysis
This notebook audits the metadata and highlights leakage risks before modeling.


```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../data/raw/Alzheimer_Dataset_Details.csv')
df.head()
```

```python
df.shape, df.isna().sum(), df['Disease'].value_counts(), df['Split'].value_counts()
```

## Leakage audit
`Filename` and `Full Path` are excluded because they encode disease-related naming or folders. Constant metadata columns are also excluded.


```python
for column in ['Extension','Width','Height','Channels','Color Mode']:
    print(column, df[column].nunique(), df[column].unique()[:5])
```
